```python
# lliberies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# paht directori
path = "Fets_delictius_i_infraccions_administratives_de_l__mbit_de_l_odi_i_discriminaci__20240517.csv"

# carreguem el dataset
data = pd.read_csv(path, sep=',')

# mostrem les primeres files
print(data.head())
```

```python
# mostrem la informació del dataset
print(data.info())
```

```python
# mostrem nulls
print(data.isnull().sum())
```

```python
# descartem columnes que no ens interessen: Nombre víctimes, Canal dels fets. 
df = data.drop(['Tipus de lloc dels fets', 'Tipus de fet', 
                'Títol del fet (Codi Penal) o Llei de referència', 
                'Àmbit fet','Nombre víctimes', 'Canal dels fets'], axis=1)

# mostrem les primeres files
print(df.head())
```

```python
# mostrem descriptius
print(df.describe())
```

```python
df.columns
```

```python
# Agrupar les dades per municipi i any
df_grouped = df.groupby(['Municipi', 'Any']).agg({
    'Nombre fets o infraccions': 'sum',
    'Regió Policial (RP)': 'first',
    'Àrea Bàsica Policial (ABP)': 'first',
    'Província': 'first',
    'Comarca': 'first',
}).reset_index()

# mostrem les primeres files
print(df_grouped.head())
```

```python
df_grouped_tipus = df.groupby(
    ['Municipi', 'Any', 'Tipus de fet (Codi penal) o d\'Infracció (altres lleis)']
).agg(
    {
        'Nombre fets o infraccions': 'sum', # Sumar "Nombre fets o infraccions"
        'Regió Policial (RP)': 'first',
        'Àrea Bàsica Policial (ABP)': 'first',
        'Província': 'first',
        'Comarca': 'first'
    }
).reset_index()

# Mostrem el dataframe resultant
print(df_grouped_tipus.head())
```

https://analisi.transparenciacatalunya.cat/Urbanisme-infraestructures/Municipis-Catalunya-Geo/9aju-tpwc/about_data

```python
# carreguem el dataset de municipis
path_municipis = "municipis_geo.csv"
municipis = pd.read_csv(path_municipis, sep=',')

# mostrem les primeres files
print(municipis.head())
```

```python
# afegim les coordenades al dataset
df_geo = df_grouped.merge(municipis[['Nom', 'Longitud', 'Latitud']], left_on='Municipi', right_on='Nom')

# mostrem les primeres files
print(df_geo.head())
```

```python
# afegim les coordenades al dataset
df_geo_tipus = df_grouped_tipus.merge(municipis[['Nom', 'Longitud', 'Latitud']], left_on='Municipi', right_on='Nom')

# mostrem les primeres files
print(df_geo_tipus.head())
```

```python
# mostrem nulls
print(df_geo.isnull().sum())

# eliminem la columna Nom
df_geo = df_geo.drop(['Nom'], axis=1)

# mostrem les primeres files
print(df_geo.head())
```

```python
# eliminem la columna Nom
df_geo_tipus = df_geo_tipus.drop(['Nom'], axis=1)

# mostrem les primeres files
print(df_geo_tipus.head())
```

```python
# modificació de noms de columnes
df_geo = df_geo.rename(columns={'Nombre fets o infraccions': 'Fets',
                                'Regió Policial (RP)': 'Regió_Policial',
                                'Àrea Bàsica Policial (ABP)': 'Àrea_Bàsica_Policial'})

# mostrem les primeres files
print(df_geo.head())
```

```python
# modificació de noms de columnes
df_geo_tipus = df_geo_tipus.rename(columns={'Tipus de fet (Codi penal) o d\'Infracció (altres lleis)': 'Tipus de Fet',
                                            'Nombre fets o infraccions': 'Fets',
                                            'Regió Policial (RP)': 'Regió_Policial',
                                            'Àrea Bàsica Policial (ABP)': 'Àrea_Bàsica_Policial'})

# mostrem les primeres files
print(df_geo_tipus.head())
```

```python
# guardamos el dataset en un csv
df_geo.to_csv('fets_geo.csv', index=False)
df_geo_tipus.to_csv('fets_geo_tipus.csv', index=False)
```

```python
# generem TimeSliderChoropleth
import folium
from folium.plugins import TimeSliderChoropleth

# generem json
import json

def generar_json(dades):
    features = []
    for _, row in dades.iterrows():
        feature = {
            'type': 'Feature',
            'geometry': {
                'type': 'Point',
                'coordinates': [row['Longitud'], row['Latitud']]
            },
            'properties': {
                'time': row['Any'],
                'Fets': row['Fets']
            }
        }
        features.append(feature)
    return {'type': 'FeatureCollection', 'features': features}

```

```python
# generem el json
geojson = generar_json(df_geo)

# afegim el json al mapa
mapa = folium.Map(location=[41.3887901, 2.1589899], zoom_start=8)

TimeSliderChoropleth(
    data=geojson,
    styledict={
        str(year): {
            'color': 'red',
            'opacity': 0.6
        } for year in df_geo['Any'].unique()
    }
).add_to(mapa)

# Afegir control de capes
folium.LayerControl().add_to(mapa)

# Guardar el mapa com un fitxer HTML
mapa.save('mapa_infraccions.html')

# mostrem el mapa
mapa
```

```python
from folium.plugins import HeatMap, HeatMapWithTime

# Crear una llista de llistes per al HeatMap amb el format [latitud, longitud, valor]
heat_data = [[row['Latitud'], row['Longitud'], row['Fets']] for index, row in df_geo.iterrows()]

# Crear el mapa amb Folium
heat_map = folium.Map(location=[41.3851, 2.1734], zoom_start=8)  # Localització inicial al centre de Catalunya

# Afegir el HeatMap al mapa
HeatMap(heat_data).add_to(heat_map)

# Afegir marcadors amb tooltip per mostrar les dades al passar el ratolí
for idx, row in df_geo.iterrows():
    folium.CircleMarker(
        location=[row['Latitud'], row['Longitud']],
        radius=4,
        color='dark',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6,
        tooltip=f"{row['Municipi']} ({row['Any']}): {row['Fets']} infraccions"
    ).add_to(heat_map)

# Guardar el mapa com un fitxer HTML
heat_map.save('mapa_infraccions_calor.html')
```
